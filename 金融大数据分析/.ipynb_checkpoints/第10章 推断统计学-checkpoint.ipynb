# 随机数

```python
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# rand 函数返回[0,1)内的随机数，个数由参数决定。
npr.rand(10)
```

```python
npr.rand(5,5)
```

```python
# 如果生成[5,10)内的随机数
a = 5
b = 10
npr.rand(10)*(b-a)+a
```

```python
# 多维数组
npr.rand(5,5)*(b-a) +a
```

```python
sample_size = 500
# 指定组成的随机数
rn1 = npr.rand(sample_size, 3)
# 从low到hight的随机数
rn2 = npr.randint(0, 10, sample_size)
# 半开区间[0.0,1.0)内的随机浮点数
rn3 = npr.sample(size=sample_size)
a = [0,25,50,75, 100]
# 给定一维数组中的随机样本
rn4 = npr.choice(a, size=sample_size)
```

```python
fig,((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(7,7))
ax1.hist(rn1, bins=25, stacked=True)
ax1.set_title('rand')
ax1.set_ylabel('frequency')
ax1.grid(True)

ax2.hist(rn2, bins=25)
ax2.set_title('randint')
ax2.grid(True)

ax3.hist(rn3, bins=25)
ax3.set_title('sample')
ax3.set_ylabel('frequency')
ax3.grid(True)

ax4.hist(rn4, bins=25)
ax4.set_title('choice')
ax4.grid(True)
```

```python
"""
1. 均值为0，标准差为1的标准正太分布
2. 均值为100， 标准差为20的正太分布
3. 自由度为0.5的卡方分布
4. γ值为1的泊松分布
"""
sample_size = 500
rn1 = npr.standard_normal(sample_size)
rn2 = npr.normal(100, 20, sample_size)
rn3 = npr.chisquare(df=0.5, size=sample_size)
rn4 = npr.poisson(lam=1.0, size=sample_size)
```

```python
fig, ((ax1, ax2,), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(7,7))
ax1.hist(rn1, bins=25)
ax1.set_title('standard normal')
ax1.set_ylabel('frquency')
ax1.grid(True)

ax2.hist(rn2, bins=25)
ax2.set_title('normal(100, 200)')
ax2.grid(True)

ax3.hist(rn3, bins=25)
ax3.set_title('chi square')
ax3.grid(True)

ax4.hist(rn4, bins=25)
ax4.set_title('Poisson')
ax4.grid(True)
```

# 模拟

## 随机变量

```python
"""
ST : T日的指数水平
r: 恒定无风险对其利率
 : S的恒定波动率
 z: 标准正太分布随机变量
"""
S0 = 100 # initial value
r=0.05 # constant short rate
sigma= 0.25 # constant volatility
T=2.0  # in years
I=10000
ST1 = S0 * np.exp((r - 0.5*sigma**2)*T + sigma*np.sqrt(T)*npr.standard_normal(I))
```

```python
plt.hist(ST1, bins=50)
plt.xlabel('index level')
plt.ylabel('frequency')
plt.grid(True)
```

```python
# 使用 lognormal 函数直接得出随机变量值。
# 提供均值，标准差
ST2 = S0 * npr.lognormal((r-0.5*sigma**2)*T, sigma * np.sqrt(T), size=I)
```

```python
plt.hist(ST2, bins=50)
plt.xlabel('index level')
plt.ylabel('frequency')
plt.grid(True)
```

```python
# 使用scipy.stats 子库 和下面的助手函数 print_statistics 比较模拟结果的分布特性
import scipy.stats as scs
def print_statistics(a1, a2):
    ''' Prints selected statistics.
    
    Parameters
    ==========
    a1, a2 : ndarray objects
        results object from simulation
    '''
    sta1 = scs.describe(a1)
    sta2 = scs.describe(a2)
    print("%14s %14s %14s" % 
        ('statistic', 'data set 1', 'data set 2'))
    print(45 * "-")
    print("%14s %14.3f %14.3f" % ('size', sta1[0], sta2[0]))
    print("%14s %14.3f %14.3f" % ('min', sta1[1][0], sta2[1][0]))
    print("%14s %14.3f %14.3f" % ('max', sta1[1][1], sta2[1][1]))
    print("%14s %14.3f %14.3f" % ('mean', sta1[2], sta2[2]))
    print("%14s %14.3f %14.3f" % ('std', np.sqrt(sta1[3]), np.sqrt(sta2[3])))
    print("%14s %14.3f %14.3f" % ('skew', sta1[4], sta2[4]))
    print("%14s %14.3f %14.3f" % ('kurtosis', sta1[5], sta2[5]))

print_statistics(ST1, ST2)
```

## 随机过程

### 几何布朗运动
[![WX20181119-144756-2x.png](https://i.postimg.cc/ZRgJrFLJ/WX20181119-144756-2x.png)](https://postimg.cc/VrqywtTh)

```python
I = 10000
M = 50
dt = T / M
S0 = 100 # initial value
S = np.zeros((M + 1, I))
S[0] = S0
for t in range(1, M + 1):
    S[t] = S[t - 1] * np.exp((r - 0.5 * sigma ** 2) * dt 
            + sigma * np.sqrt(dt) * npr.standard_normal(I))
```

```python
# 到期日的模拟几何布朗运动
plt.hist(S[-1], bins = 50)
plt.xlabel('index level')
plt.ylabel('frequency')
plt.grid(True)
```

```python
print_statistics(S[-1], ST2)
```

```python
# 前10条模拟路径
plt.plot(S[:, :10], lw=1.5)
plt.xlabel('time')
plt.ylabel('index level')
plt.grid(True)
```

### 平方根扩散
[![WX20181119-150635-2x.png](https://i.postimg.cc/DZHKWrrL/WX20181119-150635-2x.png)](https://postimg.cc/bDx5W2Fw)

```python
x0 = 0.05
kappa = 3.0
theta = 0.02
sigma = 0.1
```

```python
I = 10000
M = 50
dt = T / M
def srd_euler():
    xh = np.zeros((M + 1, I))
    x1 = np.zeros_like(xh)
    xh[0] = x0
    x1[0] = x0
    for t in range(1, M + 1):
        xh[t] = (xh[t - 1]
              + kappa * (theta - np.maximum(xh[t - 1], 0)) * dt
              + sigma * np.sqrt(np.maximum(xh[t - 1], 0)) * np.sqrt(dt)  
              * npr.standard_normal(I))
    x1 = np.maximum(xh, 0)
    return x1
x1 = srd_euler()
```

```python
plt.hist(x1[-1], bins=50)
plt.xlabel('value')
plt.ylabel('frequency')
plt.grid(True)
```

```python
plt.plot(x1[:, :10], lw =1.5)
plt.xlabel('time')
plt.ylabel('index level')
plt.grid(True)
```
